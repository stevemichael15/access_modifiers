```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")

df = pd.read_csv("/kaggle/input/ai-tool-usage-by-indian-college-students-2025/Students.csv")
print(df.info())
print(df)
```

```python
#checking for null values
df.drop("State", axis=1, inplace= True)
print(df.isna().sum())
```

```python
# Visualizing the types of AI tools used by students
ai_tool_counts = df["AI_Tools_Used"].value_counts()
ai_tool_counts.plot(kind="pie", autopct="%1.1f%%", startangle=90)
plt.title("Types of AI Tools Used by Students")
plt.ylabel("")  
plt.show()
```

```python
# Plotting the distribution of students across different streams as a pie chart
stream_counts = df["Stream"].value_counts()
stream_counts.plot(kind="pie", autopct="%1.1f%%", startangle=90)
plt.title("Distribution of Streams")
plt.show()
```

```python
#Students trust in AI tools ranging from 1 to 5
stream_counts = df["Trust_in_AI_Tools"].value_counts()
stream_counts.plot(kind="bar")
plt.title("Trust in AI")
plt.xlabel("Trust range")
plt.ylabel("No. of Students")
plt.show()
```
